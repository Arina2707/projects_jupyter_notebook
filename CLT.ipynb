## Sample generation and graphics

```python
import pandas as pd
import numpy as np
import scipy.stats as sts
import matplotlib.pyplot as plt
from math import sqrt
```

```python
%matplotlib inline
```

```python
#exponential distribution
lambd=1
exp_rv=sts.expon(scale=1)
sample=exp_rv.rvs(1000)
```

```python
x=np.linspace(0, 7, 100)
pdf=exp_rv.pdf(x)
plt.hist(sample, density=True, label="Histogram")
plt.plot(x, pdf, label="Theoretical pdf")
plt.xlim(0, 7)
plt.legend()
```

## Sorting through n

```python
#n=5
arr1=[]

#строим гистограмму выборочных средних
for i in range(1,1001):
    exp_rv1=sts.expon(scale=1)
    arr1.append(np.mean(exp_rv1.rvs(5)))

plt.hist(arr1, density="True", label="Histogram")

# строим нормальное распределние, мат. ож-ие равно 1, среднеквадрат-ое отклонение равно корень из 1/5
norm_rv=sts.norm(1, sqrt(0.2))
x1=np.linspace(0, 4, 100)
pdf1=norm_rv.pdf(x1)
plt.plot(x1, pdf1, label="Theoretical pdf")
plt.legend()
#Видно, что существует некая разница
```

```python
#n=10
arr2=[]

#строим гистограмму выборочных средних
for i in range(1,1001):
    exp_rv2=sts.expon(scale=1)
    arr2.append(np.mean(exp_rv2.rvs(10)))

plt.hist(arr2, density="True", label="Histogram")

# строим нормальное распределние, мат. ож-ие равно 1, среднеквадрат-ое отклонение равно корень из 1/10
norm_rv=sts.norm(1, sqrt(0.1))
x2=np.linspace(0, 4, 100)
pdf2=norm_rv.pdf(x2)
plt.plot(x2, pdf2, label="Theoretical pdf")
plt.legend()
#Видно, что существует некая разница, но уже лучше
```

```python
#n=50
arr3=[]

#строим гистограмму выборочных средних
for i in range(1,1001):
    exp_rv3=sts.expon(scale=1)
    arr3.append(np.mean(exp_rv3.rvs(50)))

plt.hist(arr3, density="True", label="Histogram")

# строим нормальное распределние, мат. ож-ие равно 1, среднеквадрат-ое отклонение равно корень из 1/50
norm_rv=sts.norm(1, sqrt(0.02))
x3=np.linspace(0, 4, 100)
pdf3=norm_rv.pdf(x3)
plt.plot(x3, pdf3, label="Theoretical pdf")
plt.legend()
#Разница минимальна
```
